```python
import os
import sqlite3
import pandas as pd
import requests


# Get the current working directory
cwd = os.getcwd()
#cwd += "/Vocab/"

# Specify the name of the file you want to find the path for
filename = "vocab.db"

# Use os.path.join() to construct the full path to the file
file_path = os.path.join(cwd, filename)

print("Path: ", file_path)

SQLCode = '''SELECT w.stem, w.word, w.lang, w.id, l.usage, l.book_key, b.title, b.authors
FROM WORDS w
JOIN LOOKUPS l ON w.id = l.word_key
JOIN BOOK_INFO b ON l.book_key = b.id;
'''

try:
    # Connect to the database
    conn = sqlite3.connect(file_path)
    print("Database connected: ", conn)
except:
    print("An exception occurred") 

#Tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%'",conn)

#print(Tables)

#Create dataframe
df = pd.read_sql_query(SQLCode,conn)

# Close the connection to the database
conn.close()
```

```python
df = df[df["lang"] == "en"]
df = df[df["title"] != "Dune"]
df.head()
```

```python
df2 = df.drop('book_key', axis=1)
df2 = df2.drop('id', axis=1)
df2 = df2.drop('lang', axis=1)
df2.head()
```

```python
rows = df2.shape[0]
if rows > 2000:
    df_sample = df2.sample(n=2000) #API has 1000 max requests per key. Each user can have 2 keys
else:
    df_sample = df2
```

```python
#create an account at https://dictionaryapi.com/
#access https://dictionaryapi.com/account/my-keys

# Define the two API keys
API_KEY_1 = "644443cb-bbc4-4d01-9884-bae3d87d2419"

API_KEY_2 = "644443cb-bbc4-4d01-9884-bae3d87d2419"

requests_made = 0

def get_definition(word):
    global requests_made
    # Make the API request
    api_key = API_KEY_1
    response = requests.get(f"https://www.dictionaryapi.com/api/v3/references/collegiate/json/{word}?key={api_key}")
    
    # Update the request counter
    requests_made += 1

    # Check if the response contains a definition
    if response.json() and len(response.json()) > 0 and "shortdef" in response.json()[0]:
        
        try:
            return response.json()[0]["shortdef"][0]
            
        except:
            return "No definition found"
    else:
        return "No definition found"

def get_column_definition(x):
    return get_definition(x)



# Print the definition of the word

print(get_definition("draw"))
```

```python
df_sample.reset_index(drop=True, inplace=True)
df_sample["definition"] = df_sample["stem"].apply(get_column_definition)
df_sample.head()
#you can verify your requests at https://dictionaryapi.com/account/usage-reports
```

```python
df_sample = df_sample[df_sample["definition"] != "No definition found"]
df_sample = df_sample[df_sample["definition"] != "no definition found"]
df_sample.head()
```
